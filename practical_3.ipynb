```python
mat1= np.array([(2,2,2),(4,4,4),(8,8,8)],dtype='int32')
mat2= np.array([(1,1,1),(3,3,3),(5,5,5)],dtype='int32')
print(mat1)
print()
print(mat2)
```

```python
mat1= tf.constant(mat1)
mat2= tf.constant(mat2)
print(mat1)
print(mat2)
```

```python
mat_product= tf.matmul(mat1,mat2)
mat_sum= tf.add(mat1,mat2)
print(mat_product)
print(mat_sum)
```

```python
mat3= np.array([(2,5,8),(4,8,6),(7,3,2)],dtype='float64')
print(mat3)
mat3=tf.constant(mat3)
print(mat3)
mat_det=tf.linalg.det(mat3)
print(mat_det)
```

```python
import tensorflow as tf
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

#Graph creation
x = tf.Variable(3, name="x")
y = tf.Variable(5, name="y")
f = x*x + y*y + 2

sess = tf.compat.v1.Session()
sess.run(x.initializer)
sess.run(y.initializer)
result = sess.run(f)
print(result)
sess.close()

msg = tf.constant('MACHINE LEARNING')
sess = tf.Session()
print(sess.run(msg))
```
